```python
dfg = super_clean.groupby(by=['London_Borough', 'Year']).mean()
dfg.head()
```

```python
def price_ratio(d):
    y1998 = float(d['Average_price'][d['Year']==1998])
    y2018 = float(d['Average_price'][d['Year']==2018])
    ratio = [y2018/y1998]
    return ratio
```

```python
for my_list in List_LB:
    dfg.xs(my_list,level='London_Borough')
    ratio_list[my_list] = (dfg.loc[my_list, 2018])/(dfg.loc[my_list, 1998])
  
```
